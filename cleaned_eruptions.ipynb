```python
import pandas as pd
```

```python
# CSV path

csv_path = "Eruptions1950-2021.csv"
```

```python
# Import the books.csv file as a DataFrame
eruptions_df = pd.read_csv(csv_path, encoding="utf-8")
eruptions_df.head()
```

```python
eruptions_df.dtypes
```

```python
cleaned_df =eruptions_df.drop([0])

cleaned_df
```

```python
cleaned_df = cleaned_df.drop(columns=['Search Parameters', 'Tsu','Eq'])
cleaned_df
```

```python
cleaned_df.dtypes
```

```python
import numpy as np
```

```python
df_filtered=cleaned_df.dropna(subset=['Dy'])
df_filtered
```

```python

df_filtered["Mo"] = df_filtered["Mo"].astype(int)
```

```python
df_filtered["Mo"] = df_filtered["Mo"].apply(lambda x: '{0:0>2}'.format(x))
df_filtered.head()
```

```python
df_filtered["Dy"] = df_filtered["Dy"].astype(int)
```

```python
df_filtered["Dy"] = df_filtered["Dy"].apply(lambda x: '{0:0>2}'.format(x))
df_filtered.head()
```

```python
# Push the remade DataFrame to a new CSV file
df_filtered.to_csv("cleaned_eruptions.csv",encoding="utf-8", index=False, header=True)
```
